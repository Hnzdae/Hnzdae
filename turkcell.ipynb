```python
import pandas as pd
import matplotlib.pyplot as plt

# Örnek veri seti oluşturma
veriler = {
    'Tarih': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04', '2022-01-05', '2022-01-06', '2022-01-07','2022-01-08','2022-01-09','2022-01-10','2022-01-11','2022-01-12','2022-01-13','2022-01-14','2022-01-15','2022-01-16','2022-01-17','2022-01-18','2022-01-19','2022-01-20','2022-01-21','2022-01-22','2022-01-23','2022-01-24','2022-01-25','2022-01-26','2022-01-27','2022-01-28','2022-01-29','2022-01-30'],
    'Süre': [30, 45, 60, 50, 55, 40, 65, 0, 21, 25, 24, 78, 98, 45, 23, 0, 0, 0, 4, 52, 0, 85, 55, 47, 0, 10, 15, 0, 0, 6]  # 30 tane veri
}


# Veri çerçevesi oluşturma
df = pd.DataFrame(veriler)

# Tarih sütununu tarih formatına dönüştürme
df['Tarih'] = pd.to_datetime(df['Tarih'])

# Grafiği oluşturma
plt.figure(figsize=(10, 5))
plt.plot(df['Tarih'], df['Süre'], marker='o')
plt.xlabel('Tarih')
plt.ylabel('Dinleme Süresi (Dakika)')
plt.title('Müzik Dinleme Verileri')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Grafiği gösterme
plt.show()
```
